```python
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.datasets import mnist
from keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load dữ liệu MNIST
(train_X, train_y), (test_X, test_y) = mnist.load_data()
train_X = train_X / 255.0
test_X = test_X / 255.0
train_y = to_categorical(train_y)
test_y = to_categorical(test_y)

# Xây dựng mô hình Keras (MLP)
model_keras_mnist = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model_keras_mnist.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Huấn luyện mô hình
history_keras_mnist = model_keras_mnist.fit(train_X, train_y, epochs=10, validation_data=(test_X, test_y))

# Vẽ biểu đồ kết quả
plt.plot(history_keras_mnist.history['accuracy'], label='Train Accuracy')
plt.plot(history_keras_mnist.history['val_accuracy'], label='Validation Accuracy')
plt.title('Keras MLP Model Accuracy (MNIST)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from keras.layers import Conv2D, MaxPooling2D

# Thay đổi hình dạng dữ liệu
train_X_cnn = train_X.reshape(-1, 28, 28, 1)
test_X_cnn = test_X.reshape(-1, 28, 28, 1)

# Xây dựng mô hình CNN
model_cnn_mnist = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model_cnn_mnist.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Huấn luyện mô hình
history_cnn_mnist = model_cnn_mnist.fit(train_X_cnn, train_y, epochs=10, validation_data=(test_X_cnn, test_y))

# Vẽ biểu đồ kết quả
plt.plot(history_cnn_mnist.history['accuracy'], label='Train Accuracy')
plt.plot(history_cnn_mnist.history['val_accuracy'], label='Validation Accuracy')
plt.title('CNN Model Accuracy (MNIST)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from keras.layers import SimpleRNN

# Xây dựng mô hình RNN
model_rnn_mnist = Sequential([
    SimpleRNN(128, input_shape=(28, 28), return_sequences=True),
    SimpleRNN(64, return_sequences=True),
    SimpleRNN(32),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model_rnn_mnist.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Huấn luyện mô hình
history_rnn_mnist = model_rnn_mnist.fit(train_X, train_y, epochs=10, validation_data=(test_X, test_y))

# Vẽ biểu đồ kết quả
plt.plot(history_rnn_mnist.history['accuracy'], label='Train Accuracy')
plt.plot(history_rnn_mnist.history['val_accuracy'], label='Validation Accuracy')
plt.title('RNN Model Accuracy (MNIST)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
