```python
# Tự sinh tập DATA với 1000 sinh viên 
import numpy as np
import pandas as pd

# Số lượng sinh viên mỗi môn
num_students_mon1 = 500
num_students_mon2 = 500

# Hàm để sinh điểm dựa trên tỷ lệ yêu cầu
def generate_scores(ratios, num_students):
    scores = []
    for score, ratio in enumerate(ratios, start=1):
        scores.extend([score] * int(ratio * num_students))
    np.random.shuffle(scores)
    return scores

# Tạo điểm cho Môn 1 với tỷ lệ (10%, 10%, 20%, 60%)
ratios_mon1 = [0.1, 0.1, 0.2, 0.6]
scores_mon1 = generate_scores(ratios_mon1, num_students_mon1)

# Tạo điểm cho Môn 2 với tỷ lệ (10%, 20%, 20%, 50%)
ratios_mon2 = [0.1, 0.2, 0.2, 0.5]
scores_mon2 = generate_scores(ratios_mon2, num_students_mon2)

# Tạo DataFrame cho dữ liệu sinh viên
data = {
    'Student_ID': [f'SV{i+1}' for i in range(num_students_mon1 + num_students_mon2)],
    'Subject': ['Môn 1'] * num_students_mon1 + ['Môn 2'] * num_students_mon2,
    'Score': scores_mon1 + scores_mon2
}

df = pd.DataFrame(data)

# Hiển thị DataFrame
print(df.head(50))

# Lưu dữ liệu vào file CSV (tùy chọn)
df.to_csv('student_scores.csv', index=False)
```
