```python
import pandas as pd
import numpy as np
import random

# Tạo dữ liệu mẫu cho 100 học sinh
def generate_student_data(num_students, error_rate=0.1):
    data = []
    
    for student_id in range(1, num_students + 1):
        # Tạo điểm bình thường (từ 0 đến 10)
        score1 = random.randint(0, 10) if random.random() > error_rate else random.choice([random.randint(-5, 10), np.nan]) 
        score2 = random.randint(0, 10) if random.random() > error_rate else random.choice([random.randint(-5, 10), np.nan]) 
        score3 = random.randint(0, 10) if random.random() > error_rate else random.choice([random.randint(-5, 10), np.nan]) 
        
        # Tính Điểm 4 (ví dụ: trung bình cộng của Điểm 1, Điểm 2 và Điểm 3)
        # score4 = (score1 + score2 + score3) / 3 if all(v is not np.nan for v in [score1, score2, score3]) else np.nan
        
        # Lưu vào danh sách dữ liệu
        data.append([student_id, score1, score2, score3])
    
    # Tạo DataFrame
    df = pd.DataFrame(data, columns=['Id', 'Điểm 1', 'Điểm 2', 'Điểm 3'])
    
    print(df)
    return df

# Tạo dữ liệu cho 100 học sinh, tỷ lệ lỗi là 10%
student_data = generate_student_data(100, error_rate=0.2)

# Lưu dữ liệu vào file CSV
file_path = 'student_scores.csv'
student_data.to_csv(file_path, index=False)

# Hiển thị đường dẫn file đã lưu
print(f"Dữ liệu đã được lưu vào file '{file_path}'")
```

```python
student_data = pd.read_csv(file_path)

# Hiển thị dữ liệu gốc
print("Dữ liệu gốc:")
print(student_data)

# Làm sạch dữ liệu
def clean_data(df):
    # Bước 1: Loại bỏ các điểm không hợp lệ (nằm ngoài phạm vi từ 0 đến 10 hoặc NaN)
    df_cleaned = df.copy()
    df_cleaned[['Điểm 1', 'Điểm 2', 'Điểm 3']] = df_cleaned[['Điểm 1', 'Điểm 2', 'Điểm 3']].applymap(lambda x: x if isinstance(x, (int, float)) and 0 <= x <= 10 else np.nan)
    
    # Bước 2: Loại bỏ học sinh thiếu điểm (tức là có NaN trong bất kỳ cột nào của các điểm)
    df_cleaned = df_cleaned.dropna(subset=['Điểm 1', 'Điểm 2', 'Điểm 3'])
    
    return df_cleaned

# Làm sạch dữ liệu
cleaned_data = clean_data(student_data)

# Hiển thị dữ liệu đã làm sạch
print("\nDữ liệu đã làm sạch:")
print(cleaned_data)

# Lưu dữ liệu đã làm sạch vào file CSV mới
file_path_cleaned = 'cleaned_student_scores.csv'
cleaned_data.to_csv(file_path_cleaned, index=False)

# Hiển thị đường dẫn file đã làm sạch
print(f"Dữ liệu đã được lưu vào file '{file_path_cleaned}'")
```

```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Đọc dữ liệu đã làm sạch (dữ liệu chỉ có 3 điểm: Điểm 1, Điểm 2, Điểm 3)
df = pd.read_csv('cleaned_student_scores.csv')

# Kiểm tra dữ liệu có bị thiếu cột 'Điểm 4' không
if 'Điểm 4' not in df.columns:
    # Tạo một cột 'Điểm 4' mới và dự đoán giá trị của nó
    X = df[['Điểm 1', 'Điểm 2', 'Điểm 3']].values
    
    # Ở đây, ta sẽ giả định rằng có giá trị thực tế cho Điểm 4 để huấn luyện mô hình
    # Bạn sẽ cần có dữ liệu thực tế cho Điểm 4 để huấn luyện, nếu không bạn cần thêm dữ liệu này
    # Dưới đây là một ví dụ giả sử rằng bạn đã có dữ liệu Điểm 4
    y = np.random.rand(len(df)) * 10  # Giả sử Điểm 4 là một giá trị ngẫu nhiên từ 0 đến 10 (thay bằng dữ liệu thực tế của bạn)

    # Chuẩn hóa dữ liệu (Bước chuẩn hóa trước khi huấn luyện mô hình)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)  # Chuẩn hóa các đặc trưng (Điểm 1, 2, 3)

    # Huấn luyện mô hình Linear Regression
    model = LinearRegression()
    model.fit(X_scaled, y)

    # Dự đoán Điểm 4 cho tất cả sinh viên
    predicted_scores = model.predict(X_scaled)

    # Thêm cột 'Điểm 4' vào DataFrame
    df['Điểm 4'] = predicted_scores

# Tạo giao diện Tkinter
def predict_score():
    # Lấy mã sinh viên từ ô nhập liệu
    student_id = int(entry_student_id.get())
    
    # Kiểm tra mã sinh viên có hợp lệ không
    if student_id not in df['Id'].values:
        messagebox.showerror("Lỗi", "Mã sinh viên không hợp lệ.")
        return
    
    # Lấy dữ liệu của sinh viên
    student_data = df[df['Id'] == student_id].iloc[0]
    
    # Dự đoán điểm 4 (nếu cần)
    student_scores = np.array([student_data['Điểm 1'], student_data['Điểm 2'], student_data['Điểm 3']]).reshape(1, -1)
    
    # Chuẩn hóa dữ liệu sinh viên
    student_scores_scaled = scaler.transform(student_scores)
    
    # Dự đoán điểm 4 cho sinh viên
    predicted_score = model.predict(student_scores_scaled)
    
    # Hiển thị kết quả
    label_result.config(text=f"Dự đoán Điểm 4: {predicted_score[0]:.2f}")

# Khởi tạo cửa sổ Tkinter
root = tk.Tk()
root.title("Dự đoán Điểm 4 của Học Sinh")

# Thêm các thành phần giao diện
label_student_id = tk.Label(root, text="Nhập mã sinh viên:")
label_student_id.pack()

entry_student_id = tk.Entry(root)
entry_student_id.pack()

button_predict = tk.Button(root, text="Dự đoán", command=predict_score)
button_predict.pack()

label_result = tk.Label(root, text="Dự đoán Điểm 4 sẽ hiển thị ở đây.")
label_result.pack()

# Chạy giao diện
root.mainloop()
```
