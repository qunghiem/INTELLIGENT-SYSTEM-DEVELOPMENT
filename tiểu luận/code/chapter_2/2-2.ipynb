```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a simple Neural Network structure
G = nx.DiGraph()

# Add nodes for layers
input_layer = ['X1', 'X2', 'X3']
hidden_layer = ['H1', 'H2', 'H3']
output_layer = ['Y']

# Add nodes to graph
G.add_nodes_from(input_layer, layer='input')
G.add_nodes_from(hidden_layer, layer='hidden')
G.add_nodes_from(output_layer, layer='output')

# Add edges between layers
for i in input_layer:
    for h in hidden_layer:
        G.add_edge(i, h)
        
for h in hidden_layer:
    for o in output_layer:
        G.add_edge(h, o)

# Draw the graph
pos = {
    'X1': (0, 1), 'X2': (0, 0.5), 'X3': (0, 0),
    'H1': (1, 1), 'H2': (1, 0.5), 'H3': (1, 0),
    'Y': (2, 0.5)
}

plt.figure(figsize=(6, 4))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight='bold')
plt.title("Neural Network Architecture")
plt.show()
```
