```python
import numpy as np
import pandas as pd
df = pd.read_csv('diemthi.csv')
features = ['diem10%.1', 'diem10%.2', 'diem20%', 'diemthi']
df = df[features]
df.info()
```

```python
#---check for null values---
print("Nulls")
print("=====")
print(df.isnull().sum())
```

```python
df.fillna(df.mean(), inplace = True) # replace NaN with the mean
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(df.corr(),annot=True)
#---get a reference to the current figure and set its size---
fig = plt.gcf()
fig.set_size_inches(8,8)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Bước 1: Đọc dữ liệu từ file CSV

# Bước 2: Tiền xử lý dữ liệu
# Chọn các cột cần thiết
X = df[['diem10%.1', 'diem10%.2', 'diem20%']]
y = df['diemthi']

# Bước 3: Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Bước 4: Tìm giá trị k tối ưu cho KNN
param_grid = {'n_neighbors': list(range(1, 21))}  # Thử các giá trị k từ 1 đến 20
knn = KNeighborsRegressor()

grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Bước 5: Lấy giá trị k tối ưu
best_k = grid_search.best_params_['n_neighbors']
print(f"Optimal k: {best_k}")

# Huấn luyện mô hình với giá trị k tối ưu và đánh giá hiệu suất
best_knn = KNeighborsRegressor(n_neighbors=best_k)
best_knn.fit(X_train, y_train)
y_pred = best_knn.predict(X_test)

# Đánh giá mô hình
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error with k={best_k}: {mse:.2f}")
```

```python
import pickle
#---save the model to disk---
filename = 'diemthi_model.sav'
#---write to the file using write and binary mode---
pickle.dump(best_knn, open(filename, 'wb'))
```

```python
#---load the model from disk---
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
diem1 = 8
diem2 = 8
diem3 = 10
prediction = loaded_model.predict([[diem1, diem2, diem3]])
print(prediction)
```
